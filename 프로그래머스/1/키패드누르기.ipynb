### 거리 계산
무조건 1칸씩 이동, 대각선 이동 불가<br>
절대값 (x1 - x2) + 절대값 (y1 - y2)<br><br>
ex)L:[1,0] , R:[2,2] , Target=[3,1]<br>
L-Target : (1-3) + (0-1) = 3<br>
R-Target : (2-3) + (2-1) = 2<br>

<br><br>    
### 최종 흐름
1.초기 왼손(10), 오른손(12) 고정<br>
2.다음 누를 번호 왼손 고정(1,4,7) 혹은 오른손 고정(3,6,9) 유무 판단<br>
3.2번 해당 안되면, 왼손, 오른손, 누를 번호 좌표 계산<br>
4.비교 후 가까운 손으로, 같을경우 사용 손에 따라 해당 번호 클릭(answer+(L/R), L=num or R=num)<br>

````python
def solution(numbers, hand):
    answer=''
    L,R=10,12
    
    def LRcheck(answer,L,R,num):
        if num in [1,4,7]:
            answer+='L'
            L=num
            return answer,L,R
        if num in [3,6,9]:
            answer+='R'
            R=num
            return answer,L,R

    def get_coord(n):
        if n==0: return 3,1
        else: return (n-1)//3,(n-1)%3

    def get_dis(x1,y1,x2,y2):
        return abs(x2-x1)+abs(y2-y1)

    def check(dl,dr,answer,L,R,num):
        if dl>dr or ((dl==dr) and (hand[0]=='r')):
            answer+='R'
            R=num
        else :
            answer+='L'
            L=num
        return answer,L,R

    
    for num in numbers:
        if num not in [2,5,8,0]:
            answer,L,R=LRcheck(answer,L,R,num)
            
        else:
            tx,ty=get_coord(num)
            lx,ly=get_coord(L)
            rx,ry=get_coord(R)
            dl=get_dis(lx,ly,tx,ty)
            dr=get_dis(rx,ry,tx,ty)
            answer,L,R=check(dl,dr,answer,L,R,num)
            
    return answer
````

````python
numbers=[1, 3, 4, 5, 8, 2, 1, 4, 5, 9, 5]
hand='right'
solution(numbers,hand)
````
